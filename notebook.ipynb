```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('tips.csv')
```

```python
df['tips in percentage']= 100* df['tip'] / df['total_bill']
```

```python
df
```

```python
plt.figure(figsize=(12,7))


plt.subplot(1,2,1)
plt.scatter(df.index,df['total_bill'])
plt.xlabel('Index')
plt.ylabel('Total bill')
plt.title('Index vs Total bill')


plt.subplot(1,2,2)
plt.hist(df['total_bill']) 
plt.xlabel('total bill')
plt.ylabel('frequency')
plt.title(' histogram total bill')

plt.show()
```

分析10-20之间人数最多的可能性原因：人数多在2-4左右，可能是朋友聚餐，男性居多

```python
plt.figure(figsize=(12,7))
plt.subplot(1,2,2)
plt.hist(df['tips in percentage']) 
plt.xlabel('tips in percentage')
plt.ylabel('frequency')
plt.title(' histogram total bill')
plt.show()
```

花费最少的可能原因：可能在餐厅吸烟被老板讲了，不爽，所以花费最少

```python
plt.figure(figsize=(12,7))


plt.subplot(2,2,1)
plt.scatter(df['tip'],df['size'])
plt.xlabel('tip')
plt.ylabel('size')
plt.title('tip vs size')

plt.subplot(2,2,2)
plt.scatter(df['size'],df['price_per_person'])
plt.xlabel('size')
plt.ylabel('price_per_person')
plt.title('size vs price per person')

plt.subplot(2,2,3)
plt.scatter(df['total_bill'],df['tips in percentage'])
plt.xlabel('total_bill')
plt.ylabel('tips in percentage')
plt.title('total_bill vs tips in percentage')

plt.subplot(2,2,4)
plt.scatter(df['size'],df['tips in percentage'])
plt.xlabel('size')
plt.ylabel('tips in percentage')
plt.title('size vs tips in percentage')



plt.tight_layout()

plt.show()
```

tip vs size:when size=3,people pay the most tips.

size vs price_per_person:when size=2,price_per_person has the widest range,and the maximum is largest.

total_bill vs tips in percentage:when total bill is low,about 8,the tips are in the most percentage.
